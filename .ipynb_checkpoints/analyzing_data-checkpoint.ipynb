```python
import pandas as pd
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0,1.5,3,4.5,6,np.inf],labels=[1,2,3,4,5])
```

```python
import numpy as np
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index,test_index in split.split(df,df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
# Remove the income_cat column
for sett in (strat_train_set,strat_test_set):
    sett.drop("income_cat",axis=1,inplace=True)
```

```python
# Make a copy of the training dataset
df = strat_train_set.copy()
```

```python
df.plot(kind="scatter",x="longitude",y="latitude",grid=True,alpha=1,cmap="jet",c="median_house_value")
```
